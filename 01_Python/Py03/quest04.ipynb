#[QUEST 04] 파이썬을 사용하여 간단한 계산기 프로그램 만들기
###• 사용자로부터 두 개의 정수와 연산자(+,-,*,/,**)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요
###• 조건
###○ 사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다
###○ 정수가 입력될 때 까지 “잘못된 입력입니다. 정수를 입력해주세요.”를 출력하며 다시 입력받기를 시도합니다.
###○ 나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다.
###○ 사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
###○ math 모듈을 사용하여 제곱 연산(**)을 지원합니다.
###○ 계산기를 통해 계속 계산을 할 것인지 입력받습니다.

### Navigator : 강경수
### Driver : 천세문

```python
import math # math 모듈 불러오기

def input_number(message): # 입력 받는 함수 정의
  while True:
    try:
      number = int(input(message)) # 사용자 입력을 정수로 변환
      return number
    except ValueError: # 정수가 아닐 경우 예외 처리
      print("잘못된 입력입니다. 정수를 입력하세요.")

# 계산기 프로그램 반복 실행
while True:
  num1 = input_number("첫 번째 정수를 입력하세요: ") # 첫 번째 정수 입력
  num2 = input_number("두 번째 정수를 입력하세요: ") # 두 번째 정수 입력
  operator = input("연산자를 입력하세요 (+, -, *, /, **): ") # 연산자 입력

  try:
    if operator == '+': # 더하기
      result = num1 + num2
    elif operator == '-': # 빼기
      result = num1 - num2
    elif operator == '*': # 곱하기
      result = num1 * num2
    elif operator == '/': # 나누기
      if num2 == 0: # 두 번째 정수가 0인 경우 예외 처리
        raise ZeroDivisionError("0으로 나눌 수 없습니다.")
      result = num1 / num2
    elif operator == '**': # 제곱
      result = math.pow(num1, num2) # math 모듈에서 pow(제곱 연산자) 사용
    else:
      raise ValueError("지원하지 않는 연산자 입니다.") # 지원하지 않는 연산자 입력 시 예외 처리

    print(f"결과: {num1} {operator} {num2} = {result}") # 결과 출력

#except ZeroDivisionError:
#  print("0으로 나눌 수 없습니다.") # 두 번째 정수가 0인 경우 예외 처리
#except ValueError as e:
#  print(e) # 지원하지 않는 연산자 입력 시 예외 처리
#again = input("계속 계산하시겠습니까? (y/n): ") # 계속 계산 여부 입력
#if again.lower() != 'y':
#  print("계산기를 종료합니다.") # y가 아니면 종료
#  break
#
#  File "/tmp/ipykernel_1666/100631569.py", line 41
#    except ZeroDivisionError:
#                             ^
#SyntaxError: expected 'except' or 'finally' block
# 구문 오류 발생하여 Gemini 사용:
# try 블록의 범위 안에 except 문들이 try와 수직으로 정확히 맞물리게 해야됨!

  except ZeroDivisionError as e:
    print(f"오류: {e}") # 0으로 나눌 수 없음 ZeroDivisionErro
  except ValueError as e:
    print(f"오류: {e}") # 지원하지 않는 연산자 처리 ValueError
  except Exception as e: # 예상치 못한 오류 발생시 Exception
    print(f"예상치 못한 오류가 발생했습니다: {e}")

  again = input("계속 계산하시겠습니까? (y/n): ") # 계속 계산 여부 입력
  if again.lower() != 'y':
    print("계산기를 종료합니다.") # y가 아니면 종료
    break
```

## 디버깅 내용
##### SyntaxError 구문 오류가 발생하여 Gemini를 통해 오류에 대한 해답을 얻을 수 있었다. try-except 구문을 이용하였으나 except 문들이 try 구문 블록 범위 안에 할당하지 못하였음. 그렇기에 except 문들을 tap키를 눌러 try 범위에 할당하여 오류 해결하였고 출력 성공하였음.
